```python
df2 = df1.select_dtypes(include='number').fillna(0)
df2
```

```python
# 표준화 

scaler = StandardScaler()
df2 = scaler.fit_transform(df2)
```

```python
from sklearn.decomposition import PCA

# pca 사용 
pca = PCA(n_components=15)

# 학습한다
pca.fit(df2)

# 데이터를 변환한다. 
r1 = pca.transform(df2)

r1
```

```python
df3 = pd.read_csv(r'C:\Users\SAMSUNG\OneDrive\바탕 화면\파이널 프로젝트 데이터\Segment\Segment.csv')
df3
```

```python
# r1을 데이터프레임으로 변환 (컬럼 이름은 PC1 ~ PC10)
pca_df = pd.DataFrame(r1, columns=[f'PC{i+1}' for i in range(r1.shape[1])])

pca_df['Segment'] = df3['Segment']

# 미리 보기
pca_df
```

```python
le = LabelEncoder()
pca_df['Segment'] = le.fit_transform(pca_df['Segment'])
```

```python
# 입력과 결과로 나눈다. 
x = pca_df.drop('Segment', axis=1)
y = pca_df['Segment']

display(x)
display(y)
```

```python
# 차원축소 
pca = PCA(n_components=15)
pca_x = pca.fit_transform(x)
```

```python
# 교차검증을 통해 확인해본다.
model1 = XGBClassifier(slient=True, verbosity=0)
model2 = XGBClassifier(slient=True, verbosity=0)

# 교차 검증
kfold = KFold(n_splits=10, shuffle=True, random_state=1)

r1 = cross_val_score(model1, x, y, scoring='accuracy', cv=kfold)
r2 = cross_val_score(model1, pca_x, y, scoring='accuracy', cv=kfold)

print(r1.mean())
print(r2.mean())
```
