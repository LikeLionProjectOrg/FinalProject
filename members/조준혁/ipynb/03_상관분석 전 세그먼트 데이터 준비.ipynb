```python
y1_train = pd.read_parquet('open/train/1.회원정보/201807_train_회원정보.parquet')
y2_train = pd.read_parquet('open/train/1.회원정보/201808_train_회원정보.parquet')
y3_train = pd.read_parquet('open/train/1.회원정보/201809_train_회원정보.parquet')
y4_train = pd.read_parquet('open/train/1.회원정보/201810_train_회원정보.parquet')
y5_train = pd.read_parquet('open/train/1.회원정보/201811_train_회원정보.parquet')
y6_train = pd.read_parquet('open/train/1.회원정보/201812_train_회원정보.parquet')

combined_df2 = pd.concat([y1_train, y2_train, y3_train, y4_train, y5_train, y6_train], axis=0, ignore_index=True)
combined_df2
```
