```python
# 기본
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

## 1.회원정보_train_컬럼추출

```python
# 1. 모든 parquet 파일 경로 가져오기
file_list = glob.glob('open/train/1.회원정보/2018*_train_회원정보.parquet')

# 2. 데이터프레임 리스트 생성
df_list = [pd.read_parquet(f) for f in file_list]

# 3. 하나로 합치기
all_df = pd.concat(df_list, ignore_index=True)

# 4. 확인
print(all_df.shape)
all_df.head(10)
```

### 원하는 컬럼만 추출

```python
# 원하는 컬럼들 선택
selected_columns = [
    '기준년월',
    'ID',
    '이용가능여부_해외겸용_본인',
    '보유여부_해외겸용_본인',
    '수신거부여부_TM',
    '수신거부여부_메일',
    '수신거부여부_DM',
    '이용금액_R3M_신용체크',
    '이용금액_R3M_신용',
    '_1순위카드이용금액',
    '이용카드수_신용체크',
    '_2순위카드이용금액',
    '_1순위카드이용건수',
    '_2순위카드이용건수',
    '이용카드수_신용', 
]

# 해당 컬럼만 추출하여 새로운 데이터프레임 생성
filtered_df = all_df[selected_columns]

# 결과 확인
print(filtered_df.head())
```

## 1.회원정보_test_컬럼추출

```python
# 1. 모든 parquet 파일 경로 가져오기
file_list_test = glob.glob('open/test/1.회원정보/2018*_test_회원정보.parquet')

# 2. 데이터프레임 리스트 생성
df_list_test = [pd.read_parquet(f) for f in file_list_test]

# 3. 하나로 합치기
all_df_test = pd.concat(df_list_test, ignore_index=True)

# 4. 확인
print(all_df_test.shape)
all_df_test.head(10)
```

```python
# 원하는 컬럼들 선택
selected_columns = [
    '기준년월',
    'ID'
    '이용가능여부_해외겸용_본인',
    '보유여부_해외겸용_본인',
    '수신거부여부_TM',
    '수신거부여부_메일',
    '수신거부여부_DM',
    '이용금액_R3M_신용체크',
    '이용금액_R3M_신용',
    '_1순위카드이용금액',
    '이용카드수_신용체크',
    '_2순위카드이용금액',
    '_1순위카드이용건수',
    '_2순위카드이용건수',
    '이용카드수_신용', 
]

# 해당 컬럼만 추출하여 새로운 데이터프레임 생성
filtered_df_test = all_df_test[selected_columns]

# 결과 확인
print(filtered_df_test.head())
```

```python
# # CSV 파일로 저장
# filtered_df.to_csv('open/data/1.회원정보_test_filtered_data.csv', index=False, encoding='utf-8')
```
